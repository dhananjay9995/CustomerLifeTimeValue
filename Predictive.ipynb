```python
import pandas as pd
import pytimetk as ptk
```

```python
transaction_df2 = pd.read_excel("OnlineRetail.xlsx")
```

```python
transaction_df2.head()
transaction_df2.info()

transaction_df2.isna().sum()
transaction_df2.dropna(inplace=True)

(transaction_df2[['Quantity']] < 0).sum()

transaction_df2 = transaction_df2[transaction_df2['Quantity'] >= 0]

transaction_df2['InvoiceDate'] = pd.to_datetime(transaction_df2['InvoiceDate']).dt.normalize()

transaction_df2['TotalPrice'] = transaction_df2['Quantity']*transaction_df2['UnitPrice']

transaction_df2.glimpse()
```

```python
transaction_df2.head()
```

```python
days = 90
max_date = transaction_df2['InvoiceDate'].max()
```

```python
transaction_df2.drop(['StockCode','Country'],axis=1,inplace=True)
```

```python
cutoff_date = max_date-pd.to_timedelta(days,unit='d')
```

```python
cutoff_date
```

 #using query to remove data overlapping customers from train data in test data

```python
train_data = transaction_df2[transaction_df2['InvoiceDate'] < cutoff_date]
test_data= transaction_df2[transaction_df2['InvoiceDate'] >= cutoff_date].query('CustomerID in @train_data.CustomerID')
```

```python
train_data.glimpse()
train_data.head()
```

```python
test_data.glimpse()
test_data.head()
```

```python
target_data= test_data[['CustomerID','TotalPrice']].groupby('CustomerID').sum().assign(Sales_90_flag=1).rename({'TotalPrice':'Sales_90_value'},axis=1)
```

```python
target_data.glimpse()
target_data.head().sort_values(by='CustomerID')
```

```python
maxdate_train = train_data['InvoiceDate'].max()
```

```python
recency_df = train_data[['CustomerID','InvoiceDate']].groupby('CustomerID').apply(lambda x: int((x['InvoiceDate'].max() - maxdate_train)/pd.to_timedelta(1,"day"))).to_frame().set_axis(['Recency'],axis=1)
```

```python
recency_df
```

```python
frequency_df = train_data[['CustomerID','InvoiceNo']].groupby('CustomerID').count().set_axis(['Frequency'],axis=1)
```

```python
frequency_df
```

```python
monetary_df = train_data[['CustomerID','TotalPrice']].groupby('CustomerID').agg(
        {
            'TotalPrice':["sum","mean"]
        }
).set_axis(['Total_Sales_Value','Avg_Sales'],axis=1)
```

```python
train_data[['CustomerID','TotalPrice']].groupby('CustomerID').agg(
        {
            'TotalPrice':["sum","mean"]
        }
)
```

```python
monetary_df
```

```python
final_df = pd.concat([recency_df,frequency_df,monetary_df],axis=1).merge(target_data,left_index=True,right_index=True,how="left").fillna(0)
```

```python
final_df
```

```python
import pycaret.regression as reg
```

```python
reg_setup = reg.setup(data = final_df.drop('Sales_90_flag',axis=1), target='Sales_90_value')
```

```python
reg_best_models = reg.compare_models()
```

```python
reg_xgboost_model = reg.create_model('xgboost')
```

```python
reg_predict_model = reg.predict_model(reg_xgboost_model,data=final_df).sort_values('prediction_label',ascending=False)
```
